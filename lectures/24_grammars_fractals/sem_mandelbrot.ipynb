```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm

WINDX, WINDY = 900, 600
CX, CY = 600, 300
scale = 250

canvas = np.ones((WINDY, WINDX, 3), np.uint8) * 255

iter_num = 1500

for i in tqdm.tqdm(range(0, WINDX, 1)):
    for k in range(0, WINDY, 1):
        x =   (i - CX) / scale
        y = - (k - CY) / scale
        
        c = np.complex128(x, y)
        
        num = calc_runout_num(c, iter_num)
        
        #print(x, y, c, num)
        
        if (num < iter_num):
            canvas[k, i, 0] = num * 20 % 256
            canvas[k, i, 1] = num * 20 % 256
            
        else:
            canvas[k, i, 1] = 123
            canvas[k, i, 2] = 30

plt.figure(figsize = (15, 10))
plt.imshow(canvas)
```

```python
def calc_runout_num(c, max_num = 100):
    z = np.complex128(0.0, 0.0)
    
    for i in range(10):
        z = z**2 + c
        
        if (np.abs(z) > 2):
            return i
    
    return max_num

calc_runout_num(np.complex128(-1, 0))
```
