```python
import numpy as np

class Node:
    def __init__(self, key, val):
        self.key = key
        self.val = val
        self.next = None
        #print(f"{self.key} : {self.val}")
    
    def get_val(self):
        return self.val
        
    def set_next(self, new_next):
        self.next = new_next
    
    def append(self, key, val):
        if (self.next is None):
            self.next = Node(key, val)
        
        else:
            self.next.append(key, val)

# + add element (to the end)
# + print list
# + change k-th
# + add to k-th place
# - delete k-th element

# to HW:
# checking in changing k-th elem via stored list length
# list compression: 1 2 3 3 3 3 3 4 1 1 1 1 1 6 -> 1 2 3 4 1 6
# double-linked lists
# 

class Single_linked_list:
    def __init__(self):
        self.root = None
    
    def append_recursive(self, key, val):
        if (self.root is None):
            self.root = Node(key, val)
        
        else:
            self.root.append(key, val)
    
    def append_iterative(self, key, val):
        new_node = Node(key, val)
        
        if (self.root is None):
            self.root = Node(key, val)
            return
        
        curr = self.root
        
        while(curr.next is not None):
            curr = curr.next
        
        curr.next = new_node
    
    def __repr__(self):
        representation = []
        
        if (self.root is None):
            return ""
        
        curr = self.root
        
        while(curr is not None):
            representation.append(f"({curr.key} : \"{curr.val}\")")
            curr = curr.next
        
        string = " -> ".join(representation)
        
        return string
    
    # chaning value of k-th element in list
    def change_val(self, ind, new_val):
        curr = self.root # 0 -> 1 -> 2 -> 3
        
        for _ in range(ind):
            if (curr is None):
                print(f"element number {ind} out of range")
                return
            
            curr = curr.next
        
        if (curr is None):
            print(f"element number {ind} out of range")
            return
        
        curr.val = new_val
    
    def add_at_k(self, ind, key, val):
        new_elem = Node(key, val)
        
        #None root is an exceptional case
        if (self.root is None):
            if (ind != 0):
                print(f"trying to add element at spot {ind} in empty list. ignoring")
            
            else:
                self.root = new_elem
            
            return
        
        #change of the root is an exceptional case
        if (ind == 0):
            new_elem.next = self.root
            self.root = new_elem
        
        #approach spot № ind (curr will be the (ind-1)-th elem)
        curr = self.root
        
        for _ in range(ind - 1):
            if (curr is None):
                print(f"element number {ind} out of range")
                return
            
            curr = curr.next
        
        if (curr is None):
            print(f"element number {ind} out of range")
            return
        
        #insert new element
        #here curr is elem № ind-1
        
        next = curr.next
        curr.next = new_elem
        new_elem.next = next

sll = Single_linked_list()

elem_num = 4

for i in range(elem_num):
    key, val = np.random.randint(10), str(np.random.randint(10))
    
    #sll.append_recursive(key, val)
    sll.append_iterative(key, val)

print(sll) # (2 : "3") -> (4, "5") -> (6, "7")

sll.change_val(0, 15)

print(sll)

#sll.change_val(k, new_val)
```

```python
sll = Single_linked_list()

elem_num = 3

for i in range(elem_num):
    key, val = i, str(i)
    
    sll.append_iterative(key, val)

print(sll)
sll.add_at_k(4, 8, 9)
print(sll)
```
