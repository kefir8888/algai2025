### Динпрога 3

Субподрядчик Деда Мороза получил отрез ткани $n \times m$ сантиметров. Ему известно, что из куска ткани $a_i \times b_i$ можно сшить изделие стоимостью $c_i$, всего разных типов изделий $n$. Разрезать ткань можно только прямолинейными разрезами в целых координатах. Дорезать до половины и повернуть ножницы нельзя. Помогите субподрядчику найти оптимальную последовательность разрезов и максимальную стоимость изделий, которые можно изготовить из отреза ткани.

```python
# complexity:
# n * m * (n + m - 2)
# if m = \Theta(n) then O(n^3)

import numpy as np

height, width = 12, 14
A = [2, 4, 3]
B = [3, 6, 1]
C = [5, 12, 2]

# height, width = 4, 5
# A = [1, 1]
# B = [2, 3]
# C = [3, 4]

cost = np.zeros((height + 1, width + 1))

for a, b, c in zip(A, B, C):
    cost[a, b] = c
    #cost[b, a] = c

#print(cost)

for row in range(1, height + 1):
    for col in range(1, width + 1):
        #print(f"{row=}, {col=}:")
        
        # all possible horizontal cuts
        for cut_ind in range(1, row // 2 + 1):
            #print("horiz cut")
            curr_cost = cost[cut_ind, col] + cost[row - cut_ind, col]
            cost[row, col] = max(cost[row, col], curr_cost)

        # all possible vertical cuts
        for cut_ind in range(1, col // 2 + 1):
            #print(f"vert cut {cut_ind=}")
            curr_cost = cost[row, cut_ind] + cost[row, col - cut_ind]
            #print("possible update: ", curr_cost)
            cost[row, col] = max(cost[row, col], curr_cost)
        
        #print(cost[row, col])
        #print()
    
    #break

print(cost[1 : , 1 :])
```
