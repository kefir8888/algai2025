```python
import time

#1 default max for full array
#2 max for two many times
#3 custom max as a function call
#4 custom max in loop

def my_max(a, b):
    if (a > b):
        return a
    
    return b

def find_max_1(arr):
    return max(arr)

def find_max_2(arr):
    if (len(arr) == 0):
        return None
    
    m = arr[0]
    
    for el in arr[1:]:
        m = max(m, el)
    
    return m

def find_max_3(arr):
    if (len(arr) == 0):
        return None
    
    m = arr[0]
    
    for el in arr[1:]:
        m = my_max(m, el)
    
    return m

def find_max_4(arr):
    if (len(arr) == 0):
        return None
    
    m = arr[0]
    
    for el in arr[1:]:
        if (el > m):
            m = el
    
    return m

#operations
#print(start_time)

arr = [3, 5, 3, 2, 2, 1] * 100000

methods = [find_max_1, find_max_2, find_max_3, find_max_4]

#run multiple times to compare different runs
#trying different methods of max finding. Built-in is the fastest

for _ in range(5):
    for method in methods:
        start_time = time.time()

        m = method(arr)

        finish_time = time.time()

        #print(m)
        print(f"{str(method)} : {finish_time - start_time}")
    
    print()
```

```python
import matplotlib.pyplot as plt

#how does the execution time depend on input length?

arr = [3, 5, 3, 2, 2, 1] * 1000000

inp_len = []
exec_time = []

for i in range(1, 50):
    start_time = time.time()
    
    inp_arr = arr * i

    m = find_max_4(inp_arr)

    finish_time = time.time()
    
    time_elapsed = finish_time - start_time

    #print(m)
    print(f"{str(method)} : {time_elapsed}")
    
    inp_len.append(len(inp_arr))
    exec_time.append(time_elapsed)

plt.plot(inp_len, exec_time)
```

### Calculating Fibonacci numbers

$F_0 = 0$

$F_1 = 1$

$F_{k+1} = F_k + F_{k-1}$

```python
def Fib_def(k):
    if (k < 2):
        return k
    
    return Fib_def(k - 1) + Fib_def(k - 2)

times = []
fibs = []

for k in range(40):
    start_time = time.time()

    F_k = Fib_def(k)

    finish_time = time.time()
    
    time_elapsed = finish_time - start_time
    
    times.append(time_elapsed)
    fibs.append(F_k)

    print(f"F_{k} = {F_k} in {time_elapsed} seconds")
```

```python
plt.plot(times)
plt.show()
plt.plot(fibs)
```

```python
for i in range(1, len(fibs) - 2):
    print(fibs[i + 1] / fibs[i])
```

```python
import numpy as np

#Fibonacci with fast power

A = np.array([[1, 2],
              [3, 4]])

b = np.array([[5],
              [6]])

print(A @ b)
```

```python
#0 1 2 3 4 5 6 7
#0 1 1 2 3 5 8 13

A = np.array([[1, 1],
              [1, 0]])

f0 = np.array([[1],
               [0]])

k = 7

A_p = A.copy()

for i in range(k - 2):
    A_p @= A

print((A_p @ f0)[0, 0])
```
